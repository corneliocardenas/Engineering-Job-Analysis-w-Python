```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# Calculate Mean Salary

print('Total Mean Salary: ',df['salary_year_avg'].mean())
df.groupby('job_title_short')['salary_year_avg'].mean()
```

```python
# Calculate Median Salary

print('Total Median Salary: ',df['salary_year_avg'].median())
df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
# Average Hourly Salary by Job Title

df.groupby('job_title_short')['salary_hour_avg'].mean()
```

```python
# Job With Lowest Salary
df.iloc[df['salary_year_avg'].idxmin()]
```

```python
# Job Count by Country

df.groupby('job_country')['job_country'].count()
```

```python
df.iloc[df['salary_year_avg'].idxmax()]
```

```python
# Standard Deviation of Salaries

df['salary_year_avg'].std()
```
