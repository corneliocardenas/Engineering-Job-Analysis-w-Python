```python
# Minimum experience required for job position
position_experience_requirements = [1,2,3]

# Iterate over each experience requirements in the list of job positions
for x in position_experience_requirements:
  print(x)
```

```python
# Create a list of job positions
experience = [x for x in position_experience_requirements]

# The result will be a list of job positions
experience
```

```python
user_experience=2
user_experience
```

```python
# Create a list of job positions for which the user is qualified
qualified_positions = [x for x in position_experience_requirements if user_experience >= 2]
qualified_positions
```

```python
# List of senior level jobs
# Jobs and required experience (in years)
job_experience = {'Software Engineer': 2, 'Data Scientist': 3, 'Product Manager': 5, 'Data Analyst': 1}
```

```python
# Keep jobs with 3+ years experience
senior_jobs = [b for b in job_experience.keys() if job_experience[b] >= 3]
senior_jobs
```
