```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_counts = df['job_title_short'].value_counts()

job_counts
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.show()
```

```python
job_counts.plot(kind='bar')
plt.ylabel('Count of Postings')
plt.xlabel('')
plt.title('Job Postings by Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
 df[['job_posted_date', 'salary_year_avg']].dropna(subset=['salary_year_avg'])
```

```python
df.plot(x='job_posted_date', y='salary_year_avg', kind='line')
```

```python
# sort index by job posted date
df = df.set_index('job_posted_date')
df.sort_index(inplace=True)
df.head()
```

```python
df.plot(y='salary_year_avg', kind='line')
```

# Problems

```python
# Job Posting Location Bar Plot

job_counts = df['job_country'].value_counts().head(10)
job_counts.plot(kind='bar')
plt.ylabel('Count of Postings')
plt.xlabel('')
plt.title('Job Postings by Location')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Job Postings Trend Line Plot

month_posts = df.index.month.value_counts().sort_index()
month_posts.plot(kind='line')
plt.ylabel('Count of Postings')
plt.xlabel('Month')
plt.title('Job Postings by Month')
plt.show()
```

```python
# Top Job Locations

top_locations = df['job_country'].value_counts().head(10).sort_values(ascending=True)
top_locations.plot(kind='barh')
plt.xlabel('Count of Postings')
plt.ylabel('')
plt.title('Top 10 Job Locations')
plt.show()
```
