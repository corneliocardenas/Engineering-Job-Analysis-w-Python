```python
# Define a tuple of data science skill

job_skills = ("python", "sql", "statistics", "tableau")

job_skills
```

```python
# Accesing an element

job_skills[1]
```

```python
job_skills = ("python", "sql", "statistics", "tableau")

job_skills[:2]
```

```python
job_skills.append("excel")
```

```python
# Turn tuple into a list
job_skills_list = list(job_skills)

# Add skill to the list
job_skills_list.append("r")

# Convert list to a tuple
job_skills_tuple = tuple(job_skills_list)

job_skills_tuple
```

```python
job_skills_new_tuple = ("r", "postgresql")
job_skills += job_skills_new_tuple

job_skills
```

```python
job_roles_list = list(job_roles)

job_roles_list.append("Data Engineer")

job_roles_tuple = tuple(job_roles_list)

job_roles_tuple
```

```python
# Remove Job Role from tuple
job_role = ("Data Scientist", "Data Analyst", "Machine Learning Engineer")
```

```python
job_role_list = list(job_role)
job_role_list.remove("Data Scientist")
job_role_tuple = tuple(job_role_list)
job_role_tuple
```
