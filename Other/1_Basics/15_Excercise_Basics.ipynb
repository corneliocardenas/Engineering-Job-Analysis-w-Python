```python
# My skills
my_skills = ["Excel", "Python", "R", "SQL", "Tableau", "PowerBI", "Azure", "AWS", "SAP"]

job_roles = [
    {'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']},
    {'role': 'Data Scientist', 'skills': ['Python', 'R', 'Machine Learning', 'Deep Learning']},
    {'role': 'Machine Learning Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Scikit-Learn']},
    {'role': 'Data Engineer', 'skills': ['Python', 'Apache Spark', 'Hadoop', 'SQL']},
    {'role': 'Business Intelligence Analyst', 'skills': ['Python', 'SQL', 'Tableau', 'Power BI', 'Excel']},
    {'role': 'Quantitative Analyst', 'skills': ['R', 'Python', 'MATLAB', 'Statistics']},
    {'role': 'Operations Analyst', 'skills': ['Python', 'SQL', 'Data Visualization', 'Process Improvement']},
    {'role': 'Database Administrator', 'skills': ['SQL', 'Oracle', 'MySQL', 'Database Management']},
    {'role': 'AI Engineer', 'skills': ['Python', 'TensorFlow', 'PyTorch', 'Computer Vision']},
    {'role': 'Statistician', 'skills': ['R', 'SAS', 'Python', 'Statistical Modeling']}
]

# Determine which jobs you're qualified for
qualified_roles = []

for job in job_roles:
  # Initialize qualified flag
  qualified = True
  # Go through each skill in the skills key
  for skill in my_skills:
    if skill not in job["skills"]:
      qualified = False
      break
  if qualified:
    # Add the job dictionary to the qualified_jobs list
    qualified_roles.append(job["role"])

# Output qualified job roles
if qualified_roles:
  output_message = "Qualified for the following roles: " + ", ".join(qualified_roles)
  print(output_message)
else:
  print("There are no marching job roles for my skills.")
```

```python
# List of cars with required licenses
cars = [
    {'car': 'Toyota Corolla', 'licenses': ['B']},
    {'car': 'Ford F-150', 'licenses': ['B', 'C']},
    {'car': 'Tesla Model S', 'licenses': ['B']},
    {'car': 'Motorcycle', 'licenses': ['A']},
    {'car': 'Bus', 'licenses': ['D']}
]

# Your licenses
my_licenses = ["B"]

drivable_cars = []

for car in cars:
  drivable = True
  for license in my_licenses:
    if license not in car["licenses"]:
      drivable = False
      break
  if drivable:
    drivable_cars.append(car["car"])

drivable_cars

if drivable_cars:
  print("You can drive:", ", ".join(drivable_cars))
else:
  print("You can't drive any cars.")
```

```python
# Student scores
students = [
    {'name': 'Alice', 'score': 100},
    {'name': 'Bob', 'score': 100},
    {'name': 'Charlie', 'score': 100},
    {'name': 'David', 'score': 100},
    {'name': 'Eve', 'score': 100}
]

high_scoring_students = []

# TODO: Write a loop to find students who scored 80+
for student in students:
  if student["score"] >= 80:
    high_scoring_students.append(student["name"])

# Print the names of the students who passed
if high_scoring_students:
  print("The students who scored the highest were:", ", ".join(high_scoring_students))
else:
  print("None of the students got high scores")
```

```python
# Recipes and required ingredients
recipes = [
    {'name': 'Pancakes', 'ingredients': ['flour', 'eggs', 'milk']},
    {'name': 'Omelette', 'ingredients': ['eggs', 'cheese', 'butter']},
    {'name': 'Grilled Cheese', 'ingredients': ['bread', 'cheese', 'butter']},
    {'name': 'Smoothie', 'ingredients': ['milk', 'banana', 'honey']}
]

# Ingredients you have
my_ingredients = ["eggs", "milk", "flour", "butter"]

# TODO: Find recipes you can make
recipe_names = []

for recipe in recipes:
  can_make = True
  for ingredient in recipe["ingredients"]:
    if ingredient not in my_ingredients:
      can_make = False
      break
  if can_make:
    recipe_names.append(recipe["name"])

# Output the result
if recipe_names:
  print("You can make:", ", ".join(recipe_names))
else:
  print("You can't make sh#t")
```
