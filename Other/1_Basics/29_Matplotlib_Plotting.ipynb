# Problems

```python
# Job Country Bar Chart

job_countries = (
    df['job_country']
    .value_counts()
    .head(5)
    .sort_values(ascending=False)
)

plt.bar(job_countries.index, job_countries.values)
plt.show()
```

```python
# Average Salary Horizontal Bar Chart

avg_salary = (
    df.groupby('job_title_short')['salary_year_avg']
    .mean()
    .sort_values(ascending=True)
)

plt.barh(avg_salary.index, avg_salary.values)
```

```python
# Average Salary Trend Line Plot

df['job_posted_month'] = df['job_posted_date'].dt.month

avg_salary = (
    df.groupby('job_posted_month')['salary_year_avg']
    .mean()
)

plt.plot(avg_salary.index, avg_salary.values)
```
