```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
df_us
```

```python
df_us['job_posted_month'] = df_us['job_posted_date'].dt.month_name()
df_us
```

```python
us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
us_pivot.reset_index(inplace=True)
us_pivot['month_no'] = pd.to_datetime(us_pivot['job_posted_month'], format='%B').dt.month

us_pivot['month_no']
```

```python
# us_pivot.sort_values('month_no', inplace=True)
us_pivot.set_index('job_posted_month', inplace=True)
# us_pivot.drop(columns='month_no', inplace=True)
```

```python
us_pivot
```

```python
us_pivot.plot(kind='line')
```

```python
top_3 = df_us['job_title_short'].value_counts().head(3)
```

```python
top_3 = top_3.index.tolist()
```

```python
us_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

```python
## Skill count per month for data analysis
df_us = df[df['job_country'] == 'United States'].copy()
```

```python
# Extract the month name from 'job_posted_date'01_Pandas_Accessing_Data.ipynb
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
```

```python
us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
us_pivot
```

```python
us_pivot = us_pivot.reset_index()
us_pivot['month_no'] = pd.to_datetime(us_pivot['job_posted_month'], format='%B').dt.month
us_pivot = us_pivot.sort_values('month_no')
us_pivot = us_pivot.set_index('job_posted_month')
us_pivot = us_pivot.drop(columns='month_no')

us_pivot
```

```python
us_pivot.plot(kind='line')
```

```python
top_3 = df_us['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
top_3
```

```python
us_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
