```python
# Drop Missing Values

df_cleaned = df.copy()
df_cleaned = df_cleaned.dropna(how='all')
id(df_cleaned) == id(df)
```

```python
# Random Sample of DataFrame

df_cleaned.sample(frac=0.00001, replace=False)
```

```python
# Fill Missing Salaries

median_salary_year = df_cleaned['salary_year_avg'].median()
median_salary_hour = df_cleaned['salary_hour_avg'].median()

print(f"Median Yearly Salary: ", median_salary_year, "\nMedian Hourly Salary: ", median_salary_hour)
```

```python
df_cleaned['salary_year_avg'] = df_cleaned['salary_year_avg'].fillna(median_salary_year)
df_cleaned['salary_hour_avg'] = df_cleaned['salary_hour_avg'].fillna(median_salary_hour)
```
